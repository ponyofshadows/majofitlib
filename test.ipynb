# 验证多项式法的准确性

```python
import sympy as sp
from sympy import symbols, Function, diff, limit, oo, simplify, expand
import numpy as np
```

```python
B, c0, c1, c2, c3 = symbols("B,c_0,c_1,c_2,c_3")
q1, q2, sgn_q2, q = symbols("q_1,q2,sgn_q2,q")
q1 = q
q2 = -q
sgn_q2 = -1
```

```python
# c_star = symbols("c_star")
c_star = (
    c2**2 * (c1**2 + 4 * c0 * c2)
    -
    2 * c0 * c1 * c2 * c3
    +
    c0**2 * c3**2
) ** 1/2

n_numer1 = simplify(2 * c2**3 - c1*c2*c3 + c3 * (c0*c3 + sgn_q2 * c_star))
n_numer2 = simplify((2 * c2**3 - c1*c2*c3 + c3 * (c0*c3 - sgn_q2 * c_star)))
n_denom = 2 * q * (c1*c3 - c2**2) * c_star

n1 = simplify(q1 / q2 * n_numer1 / n_denom)
n2 = simplify(- n_numer2 / n_denom)

mu_numer1 = simplify(c1*c2 + c0*c3 - sgn_q2*c_star)
mu_numer2 = simplify(c1*c2 + c0*c3 + sgn_q2*c_star)
mu_denom = simplify(2 * c0 * c2)

mu1 = simplify(mu_numer1 / mu_denom)
mu2 = simplify(mu_numer2 / mu_denom)
```
