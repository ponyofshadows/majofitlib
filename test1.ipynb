# 验证多项式法的准确性

```python
import sympy as sp
from sympy import symbols, Function, diff, limit, oo, simplify, series, fraction, factor, together, collect
```

```python
B, n1, n2, mu1, mu2 = symbols("B, n_1, n_2, mu_1, mu_2")
q1, q2 = symbols("q_1,q_2")
sigma1 = q1 * n1 * mu1
sigma2 = q2 * n2 * mu2
denom1 = 1 + (mu1 * B) ** 2
denom2 = 1 + (mu2 * B) ** 2
sigma_xx =  sigma1 / denom1 + sigma2 / denom2
sigma_xy = B * (sigma1 * mu1 / denom1 + sigma2 * mu2 / denom2)

D = (sigma_xx **2 + sigma_xy **2)
rho_xx = sigma_xx / D
rho_xy = - sigma_xy / D
display(rho_xx)
display(rho_xy)
```

```python
c0 = rho_xx.subs(B, 0)
c0
```

```python
c2 = factor(simplify((diff(rho_xx/2, B , 2)).subs(B,0)))
c2
```

```python
c1 = factor(simplify((diff(rho_xy, B , 1)).subs(B,0)))
c1
```

```python
c3 = factor(simplify((diff(rho_xy/6, B , 3)).subs(B,0)))
c3
```

c0-c4的定义式几乎没问题（c1与c3在文献中写错），接下来看用c0-c4表示的$\rho_{xx}$, $\rho_{xy}$（已修正符号）是否与原来的等价

```python
rho_fit_denom = 1 + (c3/c2*B)**2
rho_xx_fit = factor(together(c0 + c2 * B**2 / rho_fit_denom))
rho_xy_fit = factor(together(c1*B + c3 * B**3 / rho_fit_denom))
```

```python
sigma1, sigma2 = symbols("sigma_1,sigma_2")
display(collect(rho_xx_fit.subs({q1: sigma1 / (n1 * mu1), q2: sigma2 / (n2 * mu2)}),[B,sigma1,sigma2]))
display(collect(rho_xy_fit.subs({q1: sigma1 / (n1 * mu1), q2: sigma2 / (n2 * mu2)}),[B,sigma1,sigma2]))
```

```python
display(simplify(rho_xx_fit.subs(B, 0)-c0))
display(simplify(diff(rho_xx_fit/2,B,2).subs(B, 0)-c2))
display(simplify(diff(rho_xy_fit,B,1).subs(B, 0)-c1))
display(simplify(diff(rho_xy_fit/6,B,3).subs(B, 0)-c3))
display(simplify(rho_xx_fit-rho_xx))
display(simplify(rho_xy_fit-rho_xy))
```

看来确实等价，所以改正后的c0-c3的表达式完全没问题。
我们通过解方程可得到准确的用c0-c3表示n1,n2,mu1,mu2的表达式（对q1*q2>0有两个，对q1*q2<0只有一个)，下面检查推导过程的正确性。

```python
display(1/c0)
display(-c1/c0**2)
display(c2/c0**3)
display(c3/c0/c2)
```

```python
display(simplify(-(c1/c0**2+1/c0*mu2)/(mu1-mu2)))
display(simplify(c1/c0**2+1/c0*mu1)/(mu1-mu2))
```

```python
display(simplify(-c2/c0-(c1*c3)/(c0*c2)))
display((simplify(c3/c2-c1/c0)))
```

```python
half_neg_b = simplify(1/2*(c3 / c2 - c1 / c0))
half_sqrt_delta = sp.sqrt(half_neg_b**2 + (c2/c0+(c1*c3)/(c0*c2)))
mu1c = simplify(half_neg_b + half_sqrt_delta)
mu2c = simplify(half_neg_b - half_sqrt_delta)
gap1 = simplify(together(mu1c+mu2c-mu1-mu2))
gap2 = simplify(mu1c*mu2c-mu1*mu2)
```

```python
display(gap1)
display(gap2)
```

验证完毕，我们现在得到了正确的表达式。
